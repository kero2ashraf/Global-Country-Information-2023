# 1) give overview of the data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\dell\OneDrive\Desktop\Data analytics files\world-data-2023.csv")
data
```

```python
columns_to_str = ["Latitude","Longitude","Calling Code"]

for column in columns_to_str:
    data[column] = data[column].astype(str)
    
columns_to_float = ['Density\n(P/Km2)', 'Agricultural Land( %)','Land Area(Km2)',
                 'Birth Rate', 'Co2-Emissions', 'Forested Area (%)',
                 'CPI', 'CPI Change (%)', 'Fertility Rate', 'Gasoline Price','GDP',
                 'Gross primary education enrollment (%)', "Armed Forces size",
                 'Gross tertiary education enrollment (%)', 'Infant mortality',
                 'Life expectancy', 'Maternal mortality ratio','Minimum wage', 
                 'Out of pocket health expenditure','Physicians per thousand', 
                 'Population','Population: Labor force participation (%)', 
                 'Tax revenue (%)','Total tax rate', 'Unemployment rate', 'Urban_population']
for column1 in columns_to_float:
    data[column1]= data[column1].astype(str)
    data[column1]= data[column1].str.replace(",","")
    data[column1]= data[column1].str.replace("$","")
    data[column1]= data[column1].str.replace("%","").astype(float)

data.info()
```

# 4) Correlation analysis

# 4.1) Heatmap : an overview of the correlation
 

```python
corr_matrix = data.corr()
plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

# 4.2 Forested Area vs CO2 emissions

```python
sns.scatterplot(data=data,x='Forested Area (%)', y='Co2-Emissions',hue="Co2-Emissions")
corr_coeff = data["Forested Area (%)"].corr(data["Co2-Emissions"])
corr_coeff_rounded = round(corr_coeff,3)
print(corr_coeff_rounded)
```

# Given a correlation coefficient of -0.014, the analysis suggests that the two variables, 'Forested Area (%)' and 'CO2-Emissions', exhibit a negligible or weak correlation. 

# 4.3) GDP vs Gress tertiary education enrollment(%)

```python
sns.scatterplot(data=data, x='GDP', y='Gross tertiary education enrollment (%)',hue='Gross tertiary education enrollment (%)')
corr_coeff = data['GDP'].corr(data['Gross tertiary education enrollment (%)'])
corr_coeff_rounded = round(corr_coeff, 3)
print(corr_coeff_rounded)
```

# Given a correlation coefficient of 0.22, the analysis suggests that the two variables, 'GDP' and 'Gross tertiary education enrollment (%)', indicates a relatively weak positive correlation.
